```python
import pandas as pd
import numpy as np
import json
import pickle
from zss import simple_distance, Node
from collections import defaultdict
from parsetrees.expr_parser import webwork_parser
from parsetrees.expr_parser import Eval_parsed
```

```python
def make_tree(etree):
    if len(etree) == 3:
        parent = etree[0]
        left = etree[1]
        right = etree[2]
        #print 'create node ', parent[0]
        node = Node((parent[0], parent[1]))
        #print 'add left child'
        node = node.addkid(make_tree(left))
        #print 'add right child'
        node = node.addkid(make_tree(right))
    elif len(etree) == 2:
        parent = etree[0]
        left = etree[1]
        #print 'create node ', parent[0]
        node = Node((parent[0], parent[1]))
        #print 'add left child'
        node = node.addkid(make_tree(left))
    else:
        #print 'create node ', etree[0]
        node = Node((etree[0], etree[1]))
    return node
```

```python
def group_by_distance(student_answers, distance_group):
    '''pass in user_part_attempts for each user
    from answer log generated from answer_from_log.py
    '''
    correct_answer = student_answers[-1][2]
    correct_parse_tree = webwork_parser.parse_webwork(correct_answer)
    if correct_parse_tree[0] == None:
        return distance_group
    correct_etree = Eval_parsed.eval_parsed(correct_parse_tree[0])
    if correct_etree == None:
        return distance_group
    correct_zsstree = make_tree(correct_etree)
    for s in student_answers[:-1]:
        answer = s[2]
        parse_tree = webwork_parser.parse_webwork(answer)
        if parse_tree[0] == None:
            print "parse error"
            continue
        if parse_tree[1] != {}:
            print "exponential problem"
            continue
        etree = Eval_parsed.eval_parsed(parse_tree[0])
        if etree == None:
            continue
        zsstree = make_tree(etree)
        dist = simple_distance(correct_zsstree, zsstree)
        if dist == 0:
            continue
        if [s[2],correct_answer] in distance_group[dist]:
            continue
        distance_group[dist].append([s[2],correct_answer])
    return distance_group
```

```python
data = pickle.load(open('Assignment10.30.13_7_poisson_rs.pg_part2,p', 'rb'))

print data[('halou', ('Assignment10.30.13', 7, 1))]
```

```python
distance_group = defaultdict(list)
for d in data:
    if data[d][-1][1] == False:
        print 'students gave up'
    else:
        distance_group = group_by_distance(data[d], distance_group)
```
